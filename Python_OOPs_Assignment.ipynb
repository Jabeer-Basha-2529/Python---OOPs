# Python OOPs Questions

1. What is Object-Oriented Programming(OOP)?
- OOP is a way of writing code where we create "objects" that have data and functions together. It helps us organize code better, reuse it, and make it easier to maintain. The main ideas are class, object, inheritance, polymorphism, encapsulation and abstraction.

2. What is a class in OOP?
- A class is like a blueprint or template. For example, if I want to make many students, I first create a "Student" class that defines what a student should have (name, roll number, marks, etc.).

3. What is an object in OOP?
An object is a real thing created from a class, Like if Student is a class, then "Ram" and "Priya" are objects of that class.

4. What is the difference between abstraction and encapsulation?
Abstraction means showing only important details and hiding complex implementation(like using a remote without knowing how it works inside).
Encapsulation means wrapping data and methods together and protecting data using private variables(using __).

5. What are dunder methods in Python?
Dunder methods are methods with double underscores(__method__). They are also called magic methods. Examples: __init__, __str__, __add__, etc. Python calls them automatically in certain situations.

6. Explain the concept of inheritance in OOP.
Inheritance means a child class can take properties and methods from a parent class. It helps in code reusability. Like Dog and Cat can inherit from Animal class.

7. What is polymorphism in OOP?
Polymorphism means "many forms". The same method name can do different things depending on which object calls it. For example, speak() method in Dog says "Bark" and in Cat says "Meow".

8. How is encapsulation achieved in Python?
By using private variables with double underscore(self.__balance). We can also use getter and setter methods to control access.

9. What is a constructor in Python?
Constructor is the __init__ method. It runs automatically when we create an object. We use it to initialize variables.

10. What are class and static methods in Python?
- Class method(@classmethod): Works with the class, first parameter is cls.
- Static method(@staticmethod): Doesn't take self or cls, just a normal function inside class.

11. What is method overloading in Python?
Python doesn't support traditional method overloading. We can achieve similar thing by giving default values to parameters or by checking types inside the method.

12. What is method overriding in OOP?
When a child class has the same method name as parent class and changes its behaviour, that's called method overriding.

13. What is a property decorator in Python?
@property allows us to use a method like an attribute. It looks clean. Wer can also use @name.setter to set values.

14. Why is polymorphism important in OOP?
It makes code more flexible and easier to extend. We can write general code that works with many different types of objects.

15. What is an abstract class in Python?
An abstract class cannot be instantiated directly. It has abstract methods that must be implemented by child classes. We use ABC and @abstractmethod from abc module

16. What are the advantages of OOP?
- Code reusability
- Better organization
- Easy to maintain and update
- Real-world modeling
- Data hiding(security)

17. What is the difference between a class variable and an instance variable?
- Class variable: Shared by all objects(defined outside __init__).
- Instance variable: Unique for each object(defined inside __init__ using self).

18. What is multiple inheritance in Python?
When a class inherits from more than one parent class. Python supports it, but we have to be careful with method resolution order.

19. Explain the purpose of __str__ and __repr__ methods in Python.
- __str__: Used for nice, readable ouput when we print the object.
- __repr__: Used for official representation, usually for developers

20. What is the significance of super() function in Python?
super() is used to call the parent class method from the child class. It's very useful in inheritance, especially with __init__.

21. 20. What is the significance of __del__ in Python?  
It's the destructor. It runs automatically when an object is deleted or goes out of scope(garbage collected).

22. What is the difference between @staticmethod and @classmethod in Python?
- @classmethod receives cls  (class) as first argument.
@staticmethod doesn't receive any automatic first argument.

23. How does polymorphism work in Python with inheritance?
When different child classes override the same method from parent, we can call that method on any child object and it will run the correct version automatically.

24. What is method chaining in Python OOP?
When multiple methods are called on the same object in one line, like obj.method1().method2().method3().

25. What is the purpose of the __call__ method in Python?
It allows us to call an object like a function. If we define  __call__ in a class, we can do obj() and it will run that method.


```python
# Practical Questions
# 1. Parent class Animal and child class Dog overriding speak()
class Animal:
  def speak(self):
    print("Some generic sound")

class Dog(Animal):
  def speak(self):
    print("Bark!")

d = Dog()
d.speak()
```

```python
# 2. Abstract class Shape
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Circle(Shape):
  def __init__(self, r):
    self.radius = r
  def area(self):
    return 3.14 * self.radius ** 2

class Rectangle(Shape):
  def __init__(self, l, w):
    self.length = l
    self.width = w
  def area(self):
    return self.length * self.width
```

```python
# 3. Multi-level inheritance: Vehicle --> Car --> EletricCar
class Vehicle:
  def __init__(self, type):
    self.type = type

class Car(Vehicle):
  def __init__(self, type, brand):
    super().__init__(type)
    self.brand = brand

class ElectricCar(Car):
  def __init__(self, type, brand, battery):
    super().__init__(type, brand)
    self.battery = battery
```

```python
# 4. Polymorphism with Bird
class Bird:
    def fly(self):
        print("Bird is flying")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies fast")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")
```

```python
# 5. Encapsulation in BankAccount
class BankAccount:
  def __init__(self, balance=0):
    self.__balance = balance

  def deposit(self, amount):
    self.__balance +=amount

  def withdraw(self, amount):
    if amount <= self.__balance:
      self.__balance -= amount
    else:
      print("Not enough balance")

  def check_balance(self):
    return self.__balance
```

```python
# 6. Runtime Polymorphism demonstration
class Instrument:
  def play(self):
    print("Instrument is playing some sound")

class Guitar(Instrument):
  def play(self):
    print("Guitar is playing strum strum...")

class Piano(Instrument):
  def play(self):
    print("Paino is playing ting ting...")

#Testing runtime polymorphism
instruments = [Guitar(), Piano(), Instrument()]

for item in instruments:
  item.play()
```

```python
# 7. MathOperations class with class method and static method.
class MathOperations:
   @classmethod
   def add_numbers(cls, a, b):
    return a + b

   @staticmethod
   def subtract_numbers(a, b):
     return a - b

# Using the methods
print(MathOperations.add_numbers(10, 5))
print(MathOperations.subtract_numbers(20, 8))
```

```python
# 8. Implement a class Person with a class method to count the total number of persons created.
class Person:
  count = 0

  def __init__(self, name):
    self.name = name
    Person.count += 2

  @classmethod
  def total_persons(cls):
    return cls.count

p1 = Person("Ram")
p2 = Person("Priya")
p3 = Person("Rahul")

print("Total persons created:", Person.total_persons())
```

```python
# 9. class Fraction with attributes numerator and denominator
class Fraction:
  def __init__(self, numerator, denominator):
    self.numerator = numerator
    self.denominator = denominator

  def __str__(self):
    return f"{self.numerator}/{self.denominator}"

f1 = Fraction(3, 4)
f2 = Fraction(5, 2)
print(f1)
print(f2)
```

```python
# 10. Operator overloading demonstration
class Vector:
  def __init__(self, x, y):
    self.x = y
    self.y = y

  def __add__(self, other):
    return Vector(self.x + other.x, self.y + other.y)

  def __str__(self):
    return f"({self.x}, {self.y})"

v1 = Vector(2,3)
v2 = Vector(5,7)
v3 = v1 + v2
print(v3)
```

```python
# 12. Student with average_grade
class Student:
  def __init__(self, name, grades):
    self.name = name
    self.grades = grades

  def average_grade(self):
    return sum(self.grades) / len(self.grades)
```

```python
# 13. Rectangle class with set_dimensions() method
class Rectangle:
    def __init__(self):
        self.length = 0
        self.width = 0

    def set_dimensions(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Using the class
rect = Rectangle()
rect.set_dimensions(10, 5)
print("Area of rectangle:", rect.area())   # Output: 50
```

```python
# 14. class Employee with a method calculate_salary()
class Employee:
    def __init__(self, name, hourly_rate):
        self.name = name
        self.hourly_rate = hourly_rate

    def calculate_salary(self, hours_worked):
        return self.hourly_rate * hours_worked

class Manager(Employee):
    def __init__(self, name, hourly_rate, bonus):
        super().__init__(name, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self, hours_worked):
        base_salary = super().calculate_salary(hours_worked)
        return base_salary + self.bonus

# Testing
emp = Employee("Ravi", 300)
print("Employee salary:", emp.calculate_salary(40))   # 12000

mgr = Manager("Anjali", 500, 5000)
print("Manager salary:", mgr.calculate_salary(40))    # 25000 (20000 + 5000 bonus)
```

```python
# 15. Product class
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity
```

```python
# 16. Abstract Animal with Cow and Sheep
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa")
```

```python
# 17. Book class
class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"{self.title} by {self.author} ({self.year_published})"
```

```python
# 18. House and Mansion(Inheritance)
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms
```
